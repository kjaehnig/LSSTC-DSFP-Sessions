```python
class Observation():## "object" and "class" are interchangable
    def __init__(self,data):
        self.data = data## attribute
    def average(self):## method
        dsum = 0
        for i,d in enumerate(self.data):
            dsum+=d
        average = dsum/(i+1)
        return average ### this is a terrible method to calculate average
```

```python
"""INSTANCES ARE NOT THE SAME THING AS OBJECTS"""
```

```python
obs1 = Observation([0,1,2])
obs2 = Observation([4,5,6])
```

```python
print("Avg 1 = {:e}; Avg 2 = {:e}".format(obs1.average(),obs2.average()))
print("Type of Avg 1 = {:}, Type of Avg 2 = {:}".format(type(obs1),type(obs2)))
```

```python
print(obs1.data)
print(obs2.data)
```

```python
"""WE CAN MAKE NEW OBJECTS BY ADDING TO EXISTING OBJECTS. THIS IS CALLED INHERITENCE."""
```

```python
class TimeSeries(Observation): ## inherits all the methods attributes of the Observation class
    def __init__(self, time, data):
        self.time = time
        Observation.__init__(self,data) ## this calls the 
        if len(self.time) != len(self.data):
            raise ValueError("Time and Data must have the same length")
    def stop_time(self):
        return self.time[-1] 
```

```python
tobs = TimeSeries([0,1,2],[3,4,5])
print(tobs)
print("Stop Time = {:e}".format(tobs.stop_time()))
print("tobs average = {:e}".format(tobs.average()))
```

```python
x = np.linspace(0,2*np.pi,1000)
y_theory = np.sinc(x)
y = y_theory + np.random.rand(1000)
```

```python
fig = plt.figure(figsize=(8,5))
ax_data = fig.add_axes([0.1,0.4,0.8,0.8]) ### [starting_x, starting_y, size_x, size_y]
ax_residual = fig.add_axes([0.1,0.1,0.8,0.3])
ax_data.plot(x,y,label='sinc(x)')
ax_data.legend()
ax_data.set_ylabel("f(x)")

ax_residual.plot(x,y-y_theory,label='residual')
ax_residual.legend()
ax_residual.set_xlabel('x')
ax_residual.set_ylabel('residual')
```
